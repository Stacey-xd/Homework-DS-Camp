DecisionTree

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import tree
import graphviz
import numpy as np

# Завантажуємо дані
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names
print('labels:', labels)
print('features:', features)

# Розділяємо на тренувальну і тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Спочатку: шукаємо найкращі параметри через GridSearchCV
param_grid = {
    'max_depth': np.arange(1, 11),
    'criterion': ['gini', 'entropy']
}

grid_search = GridSearchCV(
    DecisionTreeClassifier(random_state=20),
    param_grid,
    cv=5,
    scoring='accuracy',
    return_train_score=True
)

grid_search.fit(X_train, y_train)

print("\nBest parameters found:", grid_search.best_params_)
print("Best cross-validation score: {:.3%}".format(grid_search.best_score_))
print("Test accuracy with best parameters: {:.3%}".format(grid_search.score(X_test, y_test)))

# Тепер: будуємо фінальну модель з найкращими параметрами
best_params = grid_search.best_params_

clf = DecisionTreeClassifier(
    criterion=best_params['criterion'],
    random_state=20,
    max_depth=int(best_params['max_depth'])  # обов'язково приводимо до int
).fit(X_train, y_train)

# Виводимо точність
print("\nFinal model performance:")
print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test, y_test)))

# Візуалізація дерева рішень
graph_viz = tree.export_graphviz(clf, out_file=None,
                                 feature_names=features,
                                 class_names=labels,
                                 filled=True,
                                 rounded=False,
                                 special_characters=True)

graph = graphviz.Source(graph_viz)
graph.view(cleanup=True)
```

RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Параметри для осмисленої моделі без перенавчання
clf = RandomForestClassifier(
    n_estimators=150,          # Менше дерев
    max_depth=3,              # Обмеження глибини
    min_samples_split=5,      # Мінімум зразків для поділу
    min_samples_leaf=3,       # Мінімум зразків у листку
    max_features='sqrt',      # sqrt кількості ознак
    bootstrap=True,           # Bootstrap для різноманітності
    random_state=42,
    criterion='entropy'       # Альтернатива gini
)

# Навчання моделі
clf.fit(X_train, y_train)

# Оцінка
train_acc = accuracy_score(y_train, clf.predict(X_train))
test_acc = accuracy_score(y_test, clf.predict(X_test))
print(f"Train accuracy = {train_acc:.3%}")
print(f"Test accuracy = {test_acc:.3%}")

from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(clf, X_train, y_train, cv=5)
print("CV scores:", cv_scores)
print(f"Mean CV accuracy: {cv_scores.mean():.3%}")
print(f"Std CV accuracy: {cv_scores.std():.3%}")
```

Gradient Boosting Decision Trees (GBDT)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier

# Завантаження даних
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names

# Розбиття даних
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Створення та навчання моделі Gradient Boosting з заданими параметрами
clf = GradientBoostingClassifier(
    learning_rate=0.05,
    n_estimators=100,
    max_depth=2,
    subsample=0.8,
    random_state=0
)
clf.fit(X_train, y_train)

# Оцінка якості моделі
print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("test accuracy= {:.3%}".format(clf.score(X_test, y_test)))

# Крок 7: Визначення важливості ознак (тепер використовуємо 'clf', а не 'model')
feature_importance = clf.feature_importances_

# Підготовка DataFrame для відображення важливості ознак
feature_importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

print("\nFeature Importances:\n", feature_importance_df)

# Крок 8: Побудова графіка важливості ознак
plt.figure(figsize=(12, 8))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.title('Feature Importance in Breast Cancer Prediction (GBDT)', fontsize=14)
plt.gca().invert_yaxis()  # Найважливіші ознаки зверху
plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Параметри для RandomizedSearchCV
param_distributions = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'max_depth': [2, 3, 4],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.3],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [1, 1.5, 2]
}

random_search = RandomizedSearchCV(
    estimator=XGBClassifier(eval_metric='logloss', random_state=20),
    param_distributions=param_distributions,
    n_iter=50,  # скільки комбінацій протестувати
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=1,
    return_train_score=True
)

random_search.fit(X_train, y_train)

print("Best parameters found:", random_search.best_params_)
print("Best cross-validation score: {:.3%}".format(random_search.best_score_))
print("Test accuracy with best parameters: {:.3%}".format(random_search.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import learning_curve, cross_val_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from xgboost import XGBClassifier

# Твоя фінальна модель
clf = XGBClassifier(
    subsample=0.6,
    reg_lambda=1.5,
    reg_alpha=0.1,
    n_estimators=200,
    max_depth=2,
    learning_rate=0.2,
    gamma=0,
    colsample_bytree=0.6,
    use_label_encoder=False,
    verbosity=0,
    random_state=42
)

# Навчання
clf.fit(X_train, y_train)

# Train/Test accuracy
train_accuracy = clf.score(X_train, y_train)
test_accuracy = clf.score(X_test, y_test)
print(f"Train accuracy = {train_accuracy:.3%}")
print(f"Test accuracy = {test_accuracy:.3%}")

# === КРОК 1: Learning curve ===
train_sizes, train_scores, test_scores = learning_curve(
    clf, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42
)

train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(8, 5))
plt.plot(train_sizes, train_mean, 'o-', label="Training score")
plt.plot(train_sizes, test_mean, 'o-', label="Cross-validation score")
plt.title("Learning Curve")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend(loc="best")
plt.grid()
plt.show()

# === КРОК 2: Confusion matrix ===
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()

# === КРОК 3: Cross-validation score stability ===
cv_scores = cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean CV accuracy: {cv_scores.mean():.3%}")
print(f"Std CV accuracy: {cv_scores.std():.3%}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Завантаження даних
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
labels = cancer.target_names
features = cancer.feature_names

# Розподіл даних: спочатку на тренувальний+валідаційний і тестовий набори,
# а потім тренувальний+валідаційний розбиваємо на окремі набори.
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, random_state=20, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.1, random_state=20, stratify=y_train_full)

# Створення моделі XGBoost з оптимальними параметрами
clf = XGBClassifier(
    random_state=20,
    use_label_encoder=False,
    eval_metric='logloss',
    learning_rate=0.1,       # Поступове навчання для кращого узагальнення
    n_estimators=300,        # Більша кількість ітерацій; рання зупинка зупинить навчання, якщо нема покращення
    max_depth=3,             # Обмеження глибини дерева для контролю складності
    subsample=0.8,           # Використання підвибірки даних
    colsample_bytree=0.8,    # Використання підвибірки ознак
    reg_alpha=0.1,           # L1-регуляризація
    reg_lambda=1.0           # L2-регуляризація
)

# Навчання моделі з використанням ранньої зупинки по валідаційному набору
clf.fit(X_train, y_train,
        eval_set=[(X_val, y_val)],
        early_stopping_rounds=10,
        verbose=True)

# Оцінка моделі
print("Train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print("Validation accuracy= {:.3%}".format(clf.score(X_val, y_val)))
print("Test accuracy= {:.3%}".format(clf.score(X_test, y_test)))

# Додаткові метрики:
y_pred = clf.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=labels))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
auc = roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1])
print("AUC: {:.4f}".format(auc))
```
