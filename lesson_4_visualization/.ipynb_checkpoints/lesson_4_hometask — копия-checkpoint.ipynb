```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

# Load the dataset
file_path = "./data/Car_Insurance_Claim.xlsx"
df = pd.read_excel(file_path)

# Data Cleaning
df = df.replace("none", np.nan).infer_objects(copy=False)
# Remove unnecessary columns
df.drop(columns=["Unnamed: 0", "ID"], inplace=True, errors="ignore")

# Convert numerical columns to float
numeric_columns = ["CREDIT_SCORE", "ANNUAL_MILEAGE", "POSTAL_CODE"]
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Fill missing numerical values with median
df = df.fillna(df.median(numeric_only=True)).infer_objects(copy=False)

# Fill missing categorical values with the most frequent value
for column in df.select_dtypes(include=["object"]).columns:
    mode_value = df[column].mode()[0]
    df[column] = df[column].fillna(mode_value).infer_objects(copy=False)

# Descriptive Statistics
print("\nDataset Overview:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

# Data Visualization

# Distribution of Credit Scores
plt.figure(figsize=(10, 5))
sns.histplot(df["CREDIT_SCORE"], bins=30, kde=True, color="blue", alpha=0.7)
plt.title("Distribution of Credit Scores")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")
plt.show()

# Annual Mileage Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["ANNUAL_MILEAGE"], bins=30, kde=True, color="green", alpha=0.7)
plt.title("Annual Mileage Distribution")
plt.xlabel("Annual Mileage (miles)")
plt.ylabel("Frequency")
plt.show()

# Income Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="INCOME", hue="INCOME", palette="viridis", legend=False, order=df["INCOME"].value_counts().index)
plt.title("Income Distribution")
plt.xticks(rotation=45)
plt.ylabel("Number of Clients")
plt.show()

# Driving Experience Among Clients
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="DRIVING_EXPERIENCE", hue="DRIVING_EXPERIENCE", palette="Blues_r", legend=False, order=df["DRIVING_EXPERIENCE"].value_counts().index)
plt.title("Driving Experience Among Clients")
plt.xlabel("Driving Experience")
plt.ylabel("Number of Clients")
plt.show()

# Vehicle Ownership and Insurance Claims
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="VEHICLE_OWNERSHIP", hue="OUTCOME", palette="Set2")
plt.title("Vehicle Ownership and Insurance Claims")
plt.xlabel("Vehicle Ownership")
plt.ylabel("Number of Clients")
plt.legend(title="Insurance Claim", labels=["No", "Yes"])
plt.show()

# Credit Score vs. Income Level (Boxplot)
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="INCOME", y="CREDIT_SCORE", hue="INCOME", palette="coolwarm", legend=False)
plt.title("Credit Score Based on Income Level")
plt.xlabel("Income Level")
plt.ylabel("Credit Score")
plt.xticks(rotation=45)
plt.show()

# Scatterplot with Regression Line: Credit Score vs. Annual Mileage
plt.figure(figsize=(10, 6))
hb = plt.hexbin(df["CREDIT_SCORE"], df["ANNUAL_MILEAGE"], gridsize=40, cmap="inferno", mincnt=1,
                norm=mcolors.LogNorm(), edgecolors="k", linewidths=0.3)
plt.colorbar(hb, label="Data Density (Log Scale)")
plt.title("Hexbin (LogNorm): Credit Score vs. Annual Mileage")
plt.xlabel("Credit Score")
plt.ylabel("Annual Mileage (miles)")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Conclusions
print("\nConclusions:")
print("- Most clients have limited driving experience (0-9 years).")
print("- Income levels among clients are unevenly distributed.")
print("- There is a correlation between credit score and income level.")
print("- Clients with lower credit scores receive")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

# Set Pandas option to suppress future warnings about silent downcasting
pd.set_option('future.no_silent_downcasting', True)

# Load the dataset
file_path = "./data/Car_Insurance_Claim.xlsx"
df = pd.read_excel(file_path)

# Data Cleaning
df.replace("none", np.nan, inplace=True)
df = df.infer_objects(copy=False)  # Ensure no silent downcasting warnings

# Remove unnecessary columns
df.drop(columns=["Unnamed: 0", "ID"], inplace=True, errors="ignore")

# Convert numerical columns to float
numeric_columns = ["CREDIT_SCORE", "ANNUAL_MILEAGE", "POSTAL_CODE"]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors="coerce")

# Fill missing numerical values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing categorical values with the most frequent value
for column in df.select_dtypes(include=["object"]).columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Display descriptive statistics in a table format
import ace_tools as tools  # Ensure you're running in an environment that supports ace_tools
tools.display_dataframe_to_user(name="Summary Statistics", dataframe=df.describe())

# Data Visualization

# Distribution of Credit Scores
plt.figure(figsize=(10, 5))
sns.histplot(df["CREDIT_SCORE"], bins=30, kde=True, color="blue", alpha=0.7)
plt.title("Distribution of Credit Scores")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")
plt.show()

# Annual Mileage Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["ANNUAL_MILEAGE"], bins=30, kde=True, color="green", alpha=0.7)
plt.title("Annual Mileage Distribution")
plt.xlabel("Annual Mileage (miles)")
plt.ylabel("Frequency")
plt.show()

# Income Distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="INCOME", hue="INCOME", palette="viridis", legend=False, order=df["INCOME"].value_counts().index)
plt.title("Income Distribution")
plt.xticks(rotation=45)
plt.ylabel("Number of Clients")
plt.show()

# Driving Experience Among Clients
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="DRIVING_EXPERIENCE", hue="DRIVING_EXPERIENCE", palette="Blues_r", legend=False, order=df["DRIVING_EXPERIENCE"].value_counts().index)
plt.title("Driving Experience Among Clients")
plt.xlabel("Driving Experience")
plt.ylabel("Number of Clients")
plt.show()

# Vehicle Ownership and Insurance Claims
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="VEHICLE_OWNERSHIP", hue="OUTCOME", palette="Set2")
plt.title("Vehicle Ownership and Insurance Claims")
plt.xlabel("Vehicle Ownership")
plt.ylabel("Number of Clients")
plt.legend(title="Insurance Claim", labels=["No", "Yes"])
plt.show()

# Credit Score vs. Income Level (Boxplot)
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="INCOME", y="CREDIT_SCORE", hue="INCOME", palette="coolwarm", dodge=False)
plt.title("Credit Score Based on Income Level")
plt.xlabel("Income Level")
plt.ylabel("Credit Score")
plt.xticks(rotation=45)
plt.show()

# Scatterplot with Regression Line: Credit Score vs. Annual Mileage
plt.figure(figsize=(10, 6))
hb = plt.hexbin(df["CREDIT_SCORE"], df["ANNUAL_MILEAGE"], gridsize=40, cmap="inferno", mincnt=1,
                norm=mcolors.LogNorm(), edgecolors="k", linewidths=0.3)
plt.colorbar(hb, label="Data Density (Log Scale)")
plt.title("Hexbin (LogNorm): Credit Score vs. Annual Mileage")
plt.xlabel("Credit Score")
plt.ylabel("Annual Mileage (miles)")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# Conclusions
print("\nConclusions:")
print("- Most clients have limited driving experience (0-9 years).")
print("- Income levels among clients are unevenly distributed.")
print("- There is a correlation between credit score and income level.")
print("- Clients with lower credit scores receive more insurance payouts.")
print("- Experienced drivers (30+ years) receive fewer insurance claims.")
print("- Vehicle owners are more likely to receive claims than those using rented vehicles.")
```
